### MODEL TRAINING

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)
Y = auto_data['price']

X_train, x_test, Y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

### Linear Regresion

```python
from sklearn.linear_model import LinearRegression

#creating an estimator object
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
linear_model.score(X_train,Y_train)
```

```python
#weights for all features of dataset
linear_model.coef_
```

```python
#weights associated with particular feature.and sort to know Which features drive the price the most i.e 
#what weightage each feature has in deciding the price
predictors = X_train.columns
coef = pd.Series(linear_model.coef_,predictors).sort_values()
print(coef)
```

```python
#predict with input that is test data for us here
y_predict = linear_model.predict(x_test)
```

```python
#now plot a graph to see how close the actual value is to the predicted value, to see how good our trained model works

%pylab inline
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')

plt.legend()
plt.show()
```

```python
r_square = linear_model.score(x_test,y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

```python
#to know how much our predicted price differ from the actual prive(on an avg)
import math

math.sqrt(linear_model_mse)
```
